```python
# Import Environment Dependencies
import tweepy
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from config import api_key, api_key_secret, access_token,access_token_secret
from pprint import pprint
from datetime import date
from datetime import timedelta
from config import gkey
import requests
import urllib
from urllib.request import urlopen
import json
import gmaps
```

```python
# Get Authentication and API Handle from Tweepy

auth = tweepy.OAuthHandler(consumer_key=api_key, consumer_secret=api_key_secret)

auth.set_access_token(access_token,access_token_secret)

api = tweepy.API(auth)
```

```python
#  # function to perform data extraction
def scrape(words):
      
# Creating DataFrame using pandas
    db = pd.DataFrame(columns=['username', 'description', 'location', 'following',
                               'followers', 'totaltweets', 'retweetcount', 'text', 'hashtags'])
   
   
 # Creating a Cursor using the .Cursor() method to search through twitter for the required tweets.
    
    tweets = tweepy.Cursor(api.search_tweets,q=words, lang="en", 
             tweet_mode='extended').items(100)
    
   
    list_tweets = [tweet for tweet in tweets]
     
    # we will iterate over each tweet in the list for extracting information about each tweet
    i = 1  
    for tweet in list_tweets:
        
        username = tweet.user.screen_name
        description = tweet.user.description
        location = tweet.user.location
        following = tweet.user.friends_count
        followers = tweet.user.followers_count
        totaltweets = tweet.user.statuses_count
        retweetcount = tweet.retweet_count
        hashtags = tweet.entities['hashtags']
        
        # Retweets can be distinguished by a retweeted_status attribute,
        # in case it is an invalid reference, except block will be executed
        try:
            text = tweet.retweeted_status.full_text
        except AttributeError:
            text = tweet.full_text
       
        hashtext = list()
        for j in range(0, len(hashtags)):
            hashtext.append(hashtags[j]['text'])
          
        # Here we are appending all the extracted information in the DataFrame
       
        ith_tweet = [username, description, location, following,
                     followers, totaltweets, retweetcount, text, hashtext]
        db.loc[len(db)] = ith_tweet
    
        i = i+1
    return db
        
        
    
#     filename = 'scraped_tweets.csv'    
    # Saving the resultset as a CSV file.
#     db.to_csv(filename)

if __name__ == '__main__':
         
# Enter Keyword
    print("Enter Keyword to search for")
    words = input()

  
    db=scrape(words)
  
```

```python
#Create a Panda Dataframe with the dataframe returned in the earlier module
db_df = pd.DataFrame(db)
db_df

# Extract the top 50 tweets where the location has a value and the retweet count> 0 into separate Panda DF
no_blnk_local_df = db_df.loc[((db_df['location'] !="") & (db_df["retweetcount"] > 0)),:].sort_values('retweetcount',ascending=False).head(50)
no_blnk_local_df
```

```python
#Declare series for Location Latitude, Location Longitude and Location Names
loc_lat= []
loc_long = []
loc_name=[]

# Using the Location of the Tweets extracted above, find their matching Latitude and Longitude
# We used GMaps for this purpose

for location in no_blnk_local_df['location']:
    location_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={location}&key={gkey}"
    geo_data = requests.get(location_url).json()
    
    try:
        loc_lat.append(geo_data["results"][0]["geometry"]["location"]["lat"])
        loc_long.append(geo_data["results"][0]["geometry"]["location"]["lng"])
        loc_name.append(location)
    except IndexError:
        print("Index Out of Range")
        
```

```python
data_df = {'location' :loc_name, 'latitude': loc_lat, 'longitude':loc_long}
location_df = pd.DataFrame(data_df)
location_df
```

```python
# Create a New Panda DF which has total number of Retweets by location

grp_retweet = no_blnk_local_df.groupby(["location"])
retweet_count = grp_retweet["retweetcount"].sum()

retweet_df = pd.DataFrame({"retweetcount":retweet_count}).sort_values('retweetcount',ascending=False)
retweet_df
```

```python
# Merge the 2 Panda DFs - The one which has Lat & Long with the Total Retweets by Location
merge_retweet_df = pd.merge(location_df,retweet_df, on="location",how="left")
merge_retweet_df = merge_retweet_df.drop_duplicates()
merge_retweet_df
```

```python
# Create the components of the Heat Map - Coordinates and Weights using the final Merge_Tweet Panda
# Coordinates being Lat and Long and the Weights being the Retweet Count

coord = merge_retweet_df[["latitude","longitude"]]

popularity = merge_retweet_df["retweetcount"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(coord, weights=popularity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# Assign the marker layer to a variable
markers = gmaps.marker_layer(coord)

# Add the layer to the map
fig1.add_layer(markers)
fig1
```
